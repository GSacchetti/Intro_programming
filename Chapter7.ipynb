```python
#ISBN
def isISBN10(code):

    """
    Returns True is the argument is a string that contains a valid ISBN-10 code,
    False otherwise.
    """
    if not isinstance(code, str) or len(code) != 10:
        return False
    
    if (code[-1] != 'X' and code[-1].isalpha()):
        return False
    
    x10 = sum((n + 1) * int(digit) for n, digit in enumerate(code[:-1])) % 11
    x10 = 'X' if x10 == 10 else str(x10)
    
    return x10 == code[-1]
    
def isISBN(code, isbn13=True):
    if not isinstance(code, str):
        return False
    if not isbn13:
        return isISBN10(code)
    if len(code) != 13:
        return False
    if not code.isdigit():
        return False
    o = sum(int(code[i]) for i in range(0, len(code[:-1]), 2))
    e = sum(int(code[j]) for j in range(1, len(code[:-1]), 2))
    x13 = (10 - (o + 3*e)%10)%10
    
    return x13 == int(code[-1])

def areISBN(list_codes, isbn13=None):
    list_check = []
    if isbn13:
        for code in list_codes:
            list_check.append(isISBN(code, True))
    elif isbn13 is False:
        for code in list_codes:
            list_check.append(isISBN(code, False))
    else:
        for code in list_codes:
            if len(str(code)) == 10:
                list_check.append(isISBN(code, False))
            elif len(str(code)) == 13:
                list_check.append(isISBN(code, True))
            else:
                list_check.append(False)
    return list_check
```

```python
#Baseball
def hit(n, occupied=None):
    if occupied is None:
        occupied = []
    occupied = list(occupied)
    points = 0
    if n != 0:
        occupied = [x + n for x in occupied] + [n]
    occupied.sort()
    points = sum(1 for base in occupied if base >= 4)
    occupied = [base for base in occupied if base < 4]
    
    return points, occupied

def inning(list_n):
    if len(list_n) == 0:
        return 0,[]
    res = hit(list_n[0])
    points = res[0]
    for i, h in enumerate(list_n):
        if i!=0:
            res = hit(h, res[1])
            points += res[0]
    return points, res[1]
```

```python
#Cool serial numbers
def serial_number(serial_nb):
    serial_nb = str(serial_nb)
    assert serial_nb.isdigit(), "invalid serial number"
    assert serial_nb != '0'*len(serial_nb), "invalid serial number"
    leading_zeroes = '0'*(8-len(serial_nb))
    return leading_zeroes + serial_nb

def solid(serial_nb):
    serial_nb = serial_number(serial_nb)
    return serial_nb == serial_nb[0]*len(serial_nb)

def radar(serial_nb):
    if solid(serial_nb):
        return False
    serial_nb = serial_number(serial_nb)
    return serial_nb[:len(serial_nb)//2] == serial_nb[:(len(serial_nb)//2-1):-1]

def repeater(serial_nb):
    if solid(serial_nb):
        return False
    serial_nb = serial_number(serial_nb)
    return serial_nb[:len(serial_nb)//2] == serial_nb[len(serial_nb)//2:]

def radar_repeater(serial_nb):
    return radar(serial_nb) and repeater(serial_nb)

def numismatist(list_nbs, kind=solid):
    list_res = []
    for serial_nb in list_nbs:
        if kind(serial_nb):
            list_res.append(serial_nb)
```

```python
#Lost professions
def next_letter(v,w):
    upper_w = w.upper()
    upper_v = v.upper()
    if upper_w.count(upper_v) == 1:
        index_last_letter = upper_w.find(upper_v) + len(upper_v) - 1
        if index_last_letter != len(upper_w)-1:
            return upper_w[index_last_letter + 1]
    return ''

def extend(v, list_w):
    word = v.upper()
    step = len(word) - 1 if len(word) > 1 else 0
    for w in list_w:
        next_l = next_letter(word[-step:], w)
        if next_l == '':
            return ''
        word += next_l
    return word

def profession(list_w, length=2):
    possible_v = [list_w[0][i:i+length] for i in range(len(list_w[0]) - length + 1)]
    for pos in possible_v:
        curr_w = extend(pos, list_w[1:])
        if curr_w != '':
            return curr_w
    return ''
```

```python
#Self-inventorying arrays 
def digit_count(string):
    return tuple(string.count(str(i)) for i in range(10))

def description(count):
    count = list(count)
    return ' '.join(f"{i}{n}" if i != 0 else str(n) for n, i in enumerate(count))

def isselfinventorying(array):
    array = list(array)
    for i in range(len(array)):
        sub_string = ' '.join(array[j] for j in range(i+1))
        if array[i] != description(digit_count(sub_string)):
            return False
    return True

def isselfinventorying_2(*array):
    return isselfinventorying(array)
```

```python
#DNA motifs
def motifs(sequence, subsequence, begin=0, end=None):
    if end is None:
        end = len(sequence)
    positions = []
    pos = sequence.find(subsequence, begin, end)
    while pos != -1:
        positions.append(pos)
        pos = sequence.find(subsequence, pos + 1, end)
    return positions
```
