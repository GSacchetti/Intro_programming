```python
#Arranging apples
#A Farmer has picked apples and wants to arrange them in some crates on pallets. One crate contains 20 apples and a pallet can carry 35 crates.
#Once the farmer knows how many apples have been picked, he will also want to know

#    how many pallets he can fill to the fullest,

#    how many full crates he can fill with the remaining apples,

#    how many apples are left.

nb_apple = int(input('Nb apples ?'))

pallets_filled = (nb_apple//20)//35
crates_left = (nb_apple//20)%35
apple_left = nb_apple%20

print(pallets_filled)
print(crates_left)
print(apple_left)
```

```python
#Grain silo
#A farmer has a field of W meters wide and L meters long. This field yields C cubic meters of grain per hectare (1 hectare = 10,000 square meters). 
#The farmer has a number of cylindrical grain silos in which he stores his harvest. Each grain silo has a radius R of meters and a height H of meters.

import math
w = float(input())
l = float(input())
c = float(input())
r = float(input())
h = float(input())

field_yield = (w*l/10000)*c
volume_cylinder = math.pi*(r**2)*h

nb_full_silo = field_yield//volume_cylinder

remaining_volume = field_yield%volume_cylinder

if remaining_volume > 0:
    height_last_silo = remaining_volume / (math.pi * (r ** 2))
    print(nb_full_silo + 1)
    print(height_last_silo)
else:
    height_last_silo = volume_cylinder / (math.pi*(r**2))
    print(nb_full_silo)
    print(height_last_silo)
```

```python
#Length of human DNA
#Compute the total length of all DNA in the human body from the estimates given in the input.

base_pair_per_cell = float(input())
length_base_pair = float(input())
total_nb_cells = float(input())

length_in_meter = (base_pair_per_cell*length_base_pair*total_nb_cells)* 1e-9
length_AU = (length_in_meter*1e-3)/149597870.691
length_trip_sun = length_AU/2

print(length_in_meter)
print(length_AU)
print(length_trip_sun)
```
